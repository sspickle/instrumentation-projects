# Check data for poisson behavior

We collected some data in the lab and stored it in [poisson.csv](poisson.csv).

Let's compare the distribution of counts in this file with the theoretical poisson distribution.

```python
import matplotlib.pyplot as pl
import numpy as np
from scipy.stats import poisson
import seaborn as sns
import pandas as pd
```

```python
y=pd.read_csv('poisson.csv').data.values
mu = y.mean()
ndata = len(y)

x=poisson.rvs(size=ndata, mu=mu) # simulated poisson samples

print("mu={0:3.1f} with {1:4d} data points.".format(mu,ndata))
```

```python
sns.kdeplot(x, label="simulated data kde")
sns.kdeplot(y, label="real data kde")
n=np.arange(min(y),max(y)) # range of n values
pl.plot(n,poisson.pmf(k=n,mu=mu),label="exact poisson") # compute the exact poisson probabilities
pl.legend()
```

So this data appears to exhibit a strong poisson behavior.
